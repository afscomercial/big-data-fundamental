The business would like to predict any of the columns of data from July 30, 2021 based on the data for July 1 to July 29, 2023. So this is a predictive analysis activity.

In our case, we will predict the **volume** based on previous data.

Feature descriptions:

* symbol: the underlying stock

* date: the dataset date

* adjusted close: the closing price of the underlying stock after extended trading session

* stock price: the closing price of the underlying stock

* option symbol: string combining other fields identifying the option

* expiration: date that the option expires

* strike: the price that the option gives you the right to trade the underlying stock

* call/put: whether the option is a Call (right to buy underlying stock) or a Put (right to sell underlying stock)

* ask/bid/meanprice: market prices of the option (last offer to sell, last offer to buy, and the average of the two prices)

* iv: how fearful the market is of a market crash or severe decline in prices

* volume: number of contracts traded today

* open interest: number of contracts currently held

* greeks (delta, gamma, vega, theta, rho): group of five metrics used to analyze the risk and potential return of options contracts.

## PCA

Dimensionality reduction is a powerful statistical and machine learning approach aimed at make the data more efficient by decreasing the number of features. This process seeks to achieve an "optimal" number of dimensions, capturing the essence of the data while minimizing redundancy.

One of the most widely used methods for dimensionality reduction is feature extraction. Feature extraction condenses the data by transforming it from a higher-dimensional feature space to a lower-dimensional one, retaining the most crucial information. Principal Component Analysis (PCA) stands as the most popular technique within feature extraction. (Reduce Data Dimensionality, 2022)

Being more specific, we can take the definition for principal components from the pca library documentation (Erdogan Taskesen, 2022) saying that "Principal components are new (latent) variables that are constructed as linear combinations or mixtures of the initial variables. These combinations are in such a way that the new variables (i.e., principal components) are uncorrelated and most of the information within the initial variables is squeezed or compressed into the first components." 

To start the dimension reduction, first it is needed to import pca library:

```python
from pca import pca
```

Then a pca instance must be initialized and pass the data that needs to be fit into the PCA model, in this case, the features (X)

```python
# Initialize pca instance
model = pca()
# Fit transform
out = model.fit_transform(X)
```

Looking into the "topfeat" element from the output, the 10 best performing features are extracted based on the loadings of the Principal Components, which are readily computed.

```python
# Print the top features. The results show that strike is best, followed by stock price for iv, etc
print(out['topfeat'])
```

The most important feature is strike, followed by "stock price for iv" and theta 

On the other hand, Explained variance refers to the variance explained by each of the principal components (eigenvectors). 

In the follwing graph we have 12 variables in the dataset, and thus 12 dimensions. PCA will optimize to store maximum variance in the first PC, then in the second and so on. This plot provides insights in the amount of information or explained variance in the data. We can clearly see that the 1st PC contains almost 25% of explained variance in total. With the top 9 PCs we cover 99.81% of all variance.



```python
model.plot()
```

We can also specify the minimum % of the explained variance that it is needed to cover in order to select the minimum number of principal components as shown bellow: 

```python
model = pca(n_components=0.95)
# Fit transform
out = model.fit_transform(X)
```

```python
out
```

We can confirm that this let the dataset with nine components:

```python
pca_df = out['PC']
pca_df
```
